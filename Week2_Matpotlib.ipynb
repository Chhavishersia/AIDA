```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,5,11) #11 numbers with 0 to 5 space
y = x**2
print(x)
print(y)
```

```python
plt.plot(x,y, "r")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Plot")
```

```python
#plt.subplot(nrows,ncols,plot_number)
#(number of rows, number of columns, plot number)
plt.subplot(2,1,1) 
plt.plot(x,y,"r--") #more colour options later
plt.subplot(2,1,2)
plt.plot(y,x,"g*-"); #instead of plot.show(), just put a ;
```

```python
#create an empty canvas
fig = plt.figure()
#add axis


#tell python to set the axis
```

```python
#left is how far you wanna go from the y axis, bottom is how far from the x axis in the +ve side, 
fig = plt.figure()

ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #these are the ratios to the screen that I have, l, b, w, h
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.6])
#main axis

#larger Figure, the blue one, Axis one
ax1.plot(x,y,'b')
ax1.set_xlabel('X_label_ax1')
ax1.set_ylabel('Y_label_ax1')
ax1.set_title('Ax1 Title');

#add next axis
ax2.plot(y,x,'r')
ax2.set_xlabel('X_label_ax2')
ax2.set_ylabel('X_label_ax2')
ax2.set_title('Ax2 Title');
```

```python
#creating dataset
cars = ['AUDI', 'BMW', 'FORD', 'TESLA', 'JAGUAR', 'MERCEDES']
data = [23, 17, 35, 29, 12, 41]

#creating plot
fig = plt.figure(figsize = (1,1))
plt.pie(data, labels = cars)
plt.title("eggs are boiling")
#show the plot
plt.show()
```
